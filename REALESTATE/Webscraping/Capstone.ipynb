```python
driver=webdriver.Chrome()
```

```python
driver.get('https://vestapreferred.com/real-estate?sort=ratingweight%2Ccreated_at%20desc')
```

```python
cost=[]
bedrooms=[]
balconoy=[]
sft=[]
address=[]
```

```python
tm=driver.find_elements(By.XPATH,'//div[@class="no-wrap"]')
for i in tm:
    if i.text is None :
        cost.append("--") 
    else:
        cost.append(i.text)
DT=driver.find_elements(By.XPATH,'//div[@class="parameters"]/span[1]')
for i in DT:
    if i.text is None :
        bedrooms.append("--") 
    else:
        bedrooms.append(i.text)
G=driver.find_elements(By.XPATH,'//div[@class="parameters"]/span[2]')
for i in G:
    if i.text is None :
       balconoy.append("--") 
    else:
        balconoy.append(i.text)
TO=driver.find_elements(By.XPATH,'//div[@class="parameters"]/span[3]')
for i in TO:
    if i.text is None :
        sft.append("--") 
    else:
        sft.append(i.text)
AD=driver.find_elements(By.XPATH,'//div[@class="no-wrap small-long-address"]')
for i in AD:
    if i.text is None :
        address.append("--") 
    else:
        address.append(i.text)                  
Vesta=pd.DataFrame([])
Vesta['Cost']=cost[0:29]
Vesta['Bed_Rooms']=bedrooms[0:29]
Vesta['Balcony']=balconoy[0:29]
Vesta['SFT']=sft[0:29]
Vesta['ADDRESS']=address[0:29]
Vesta
```
